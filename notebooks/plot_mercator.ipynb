## Making the plot

Now we're ready to extract a slice from the netcdf file:

```python
%matplotlib inline
in_file='../tos_AMSRE_L3_v7_200206-201012.nc'
options=dict(corners=wp,start_date=dt.datetime(2003,4,1),stop_date=dt.datetime(2006,3,1))
data_nc,var_nc,the_times,the_lats,the_lons,sst=get_var_2D(in_file,'tos',**options)
#
# time average to get a lat/lon array
#
sst_avg=sst.mean(axis=0)

print(the_lats.shape)
print(the_lons.shape)
print(sst_avg.shape)
```

Now make the map, writing a title string that includes the first and last dates using strftime
Note that missing values are set to violet in make_plot

```python
%matplotlib inline
fig,ax,map=make_plot(the_lons,the_lats,sst_avg,vmin=301,vmax=303.5)
start_string=the_times[0].strftime("%Y-%b-%d")
stop_string=the_times[-1].strftime("%Y-%b-%d")
ax.set_title('Mean AMSR-E tropical surface temps (K) %s to %s' % (start_string,stop_string))
fig.canvas.draw()
outfile='sst_pacific.png'
fig.savefig(outfile)
```

Make a second figure by going back to the [time,lat,lon] sst array and averaging over lat/lon
We'll ignore the dependence of gridcell area on latitude and longitude to keep things simple

```python
#
# now plot the timeseries
#
sst_time=sst.mean(axis=1) #lat average
sst_time=sst_time.mean(axis=1) #lon_average
fig=plt.figure(2)
fig.clf()
ax1=fig.add_subplot(111)
ax1.plot(the_times,sst_time)
ax1.set_title('AMSR-E Warm Poool SST (K):  %s to %s' % (start_string,stop_string))
labels = ax1.get_xticklabels()
#
# rotate the tick labels to avoid crowding
#
a=[item.set_rotation(30) for item in labels]
```
